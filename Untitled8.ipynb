```python
print("                                     <*DATA LOADING*>")
import pandas as pd
a = pd.read_csv("titanic.csv")
print(a)
```

```python
print("                                            <*DATA PREPROCESSING*>")
a.isna()
```

```python
from sklearn import preprocessing
s = preprocessing.LabelEncoder()
a["Name"] =s.fit_transform(a["Name"])
a["Name"].unique()
```

```python
a["Sex"] =s.fit_transform(a["Sex"])
a["Sex"].unique()
```

```python
a["Embarked"] =s.fit_transform(a["Embarked"])
a["Embarked"].unique()
```

```python
a["Ticket"] =s.fit_transform(a["Ticket"])
a["Ticket"].unique()
```

```python
a["Cabin"] =s.fit_transform(a["Cabin"])
a["Cabin"].unique()
```

```python
a["Age"] =s.fit_transform(a["Age"])
a["Age"].unique()
```

```python
a["Fare"] =s.fit_transform(a["Fare"])
a["Fare"].unique()
```

```python
print("                                <*DATA SPLITTING*>")

x=a.iloc[:,:-1].values
print(x)
```

```python
y=a["Survived"]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20, shuffle=True)
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
print("                                   <*DATA CLASSIFICATION*>")
print("                                     LOGISTIC REGRESSION")
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
s = LogisticRegression()
s.fit(x_train, y_train)
predictions = s.predict(x_test)
print("Accuracy", accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
print("                      DECISION TREE CLASSIFIER")
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
jg = DecisionTreeClassifier()
jg.fit(x_train, y_train)
predictions = jg.predict(x_test)
print("Accuracy", accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
